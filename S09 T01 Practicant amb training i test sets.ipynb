Importación del dataset

### Nivell 2

- **Exercici 2:**

Aplica algun procés de transformació (estandarditzar les dades numèriques, crear columnes dummies, polinomis...).

Se comprueba la cantidad de datos NaN incluídos en el dataset

```python
raw_df.isna().sum()
```

Para no eliminar un número considerable de datos que nos interesan, se reemplaza el valor NaN por el valor medio de cada columna.

```python
raw_df['ArrDelay']=raw_df['ArrDelay'].fillna(raw_df['ArrDelay'].mean())
```

```python
raw_df['CarrierDelay']=raw_df['CarrierDelay'].fillna(raw_df['CarrierDelay'].mean())
```

```python
raw_df['WeatherDelay']=raw_df['WeatherDelay'].fillna(raw_df['WeatherDelay'].mean())
```

```python
raw_df['NASDelay']=raw_df['NASDelay'].fillna(raw_df['NASDelay'].mean())
```

```python
raw_df['SecurityDelay']=raw_df['SecurityDelay'].fillna(raw_df['SecurityDelay'].mean())
```

```python
raw_df['LateAircraftDelay']=raw_df['LateAircraftDelay'].fillna(raw_df['LateAircraftDelay'].mean())
```

```python
raw_df.isna().sum()
```

Tras comprobar que se han reemplazado los datos correctamente, se decide eliminar el resto de NaN

```python
no_NaN_df = raw_df.dropna()
```

```python
no_NaN_df.isna().sum()
```

Se convierten los datos categóricos usando el método Dummies.

```python
dummies_df = pd.get_dummies(no_NaN_df, columns=["UniqueCarrier", "Origin", "Dest"])
```

```python
dummies_df
```
